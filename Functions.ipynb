```python
import numpy as np
import matplotlib.pyplot as plt
import math

def plotfuncmean(f, a, b, n):
    temparr = np.linspace(a, b, 2*n)
    integ = sum(f(temparr)*((b-a)/(2*n)))
    mean = (1/(b-a))*(integ)
    plt.plot(temparr, mean * temparr**0, label = 'Mean')

def plotfuncmed(f, a, b, n):
    temparr = np.linspace(a, b, 2*n)
    funcarr = np.vstack((temparr, f(temparr)))
    sorted_arr = np.sort(funcarr, axis=1)
    median = (sorted_arr[1, n - 1] + sorted_arr[1, n]) / 2
    plt.plot(temparr, median * temparr**0, label = 'Median')

def plotfuncsort(f, a, b, n):
    temparr = np.linspace(a, b, 2*n)
    sorted_arr = np.sort(f(temparr))
    plt.plot(temparr, sorted_arr, label = 'sort(f(x))' )

def plotfuncderiv(f, a, b, n):
    temparr = np.linspace(a, b, 2*n)
    funcarr = f(temparr)
    der = np.divide(([funcarr[i+1]-funcarr[i] for i in range(len(funcarr)-1)]),((b-a)/(2*n)))
    print(der)
    plt.plot(np.linspace(a, b, len(der)), der, label = 'f\'(x))' )


f = lambda x: np.exp(-x**2)

a = math.pi*(-1/2)
b = math.pi*(1/2)
n = 10000

x_vals = np.linspace(a, b, 2*n)
plt.plot(x_vals, f(x_vals), label='f(x)')
plotfuncsort(f,a,b,n)
plotfuncmed(f,a,b,n)
plotfuncmean(f,a,b,n)
# plotfuncderiv(f, a, b, n)

plt.legend()
plt.show()


```
